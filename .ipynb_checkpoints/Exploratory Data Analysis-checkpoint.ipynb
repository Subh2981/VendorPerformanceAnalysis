```python
pd.read_sql_query("""SELECT 
    a.VendorNumber,
    a.VendorName,
    a.Brand,
    a.PurchasePrice,
    b.Volume,
    b.Price as ActualPrice,
    SUM(a.Quantity) AS Total_Quantity,
    SUM(a.Dollars) AS Total_Dollars
FROM purchases AS a
JOIN purchase_prices AS b
    ON a.Brand = b.Brand
    where a.PurchasePrice > 0
GROUP BY 
    a.VendorNumber,
    a.VendorName,
    a.Brand
    
ORDER BY Total_Dollars;""",conn)
```

```python
pd.read_sql_query("""select VendorNo,
VendorName,
Brand,
sum(SalesDollars) as Total_sales_Dollers,
sum(SalesPrice) as Total_sales_price,
sum(SalesQuantity) as total_sales_Quantity,
sum(ExciseTax) as Total_Excise_Tax
from sales
group by
    VendorNo,
    VendorName,
    Brand
order by total_sales_price    
    """,conn)
```

```python
value_sales_summery=pd.read_sql_query("""with FrieghtSummery as(
SELECT VendorNumber,sum(Freight) as Freight_cost from vendor_invoice group by VendorNumber
),
PurchaseSummery as(SELECT 
    a.VendorNumber,
    a.VendorName,
    a.Brand,
    a.PurchasePrice,
    b.Volume,
    b.Price as ActualPrice,
    SUM(a.Quantity) AS Total_Quantity,
    SUM(a.Dollars) AS Total_Dollars
FROM purchases AS a
JOIN purchase_prices AS b
    ON a.Brand = b.Brand
    where a.PurchasePrice > 0
GROUP BY 
    a.VendorNumber,
    a.VendorName,
    a.Brand),
SalesSummery as(SELECT 
    VendorNo,
    VendorName,
    Brand,
    sum(SalesDollars) as Total_sales_Dollers,
    sum(SalesPrice) as Total_sales_price,
    sum(SalesQuantity) as total_sales_Quantity,
    sum(ExciseTax) as Total_Excise_Tax
    from sales
group by
    VendorNo,
    VendorName,
    Brand)
SELECT 
    aa.VendorNumber,
    aa.VendorName,
    aa.Brand,
    aa.PurchasePrice,
    aa.Volume,
    aa.ActualPrice,
    aa.Total_Quantity,
    aa.Total_Dollars,
    bb.Total_sales_Dollers,
    bb.Total_sales_price,
    bb.Total_sales_Quantity,
    bb.Total_Excise_tax,
    cc.Freight_cost
from PurchaseSummery as aa
left join SalesSummery as bb
    on aa.VendorNumber=bb.VendorNo
        and aa.Brand=bb.Brand
left join FrieghtSummery as cc
    on aa.VendorNumber=cc.VendorNumber
order by Total_Dollars;
""",conn)
```

```python
value_sales_summery['Volume']=value_sales_summery.Volume.astype('float64')
value_sales_summery.fillna(0,inplace=True)
```

```python
value_sales_summery['VendorName']=value_sales_summery['VendorName'].str.strip()
```

```python
value_sales_summery.info()
```

```python
value_sales_summery['GrossProfit']=value_sales_summery['Total_sales_Dollers']-value_sales_summery['Total_Dollars']
```

```python
value_sales_summery['GrossProfit'].min()
```

```python
value_sales_summery['ProfitMargin']=(value_sales_summery['GrossProfit']/value_sales_summery['Total_sales_Dollers'])*100
```

```python
value_sales_summery['StockTurnover']=value_sales_summery['total_sales_Quantity']/value_sales_summery['Total_Quantity']
```

```python
value_sales_summery['SalestoPurchaseRatio']=value_sales_summery['Total_sales_Dollers']/value_sales_summery['Total_Dollars']
```

```python
cursor=conn.cursor()
```

```python
value_sales_summery.columns
```

```python
cursor.execute("""create table Vendor_Sales_Summery(
VendorNumber int,
VendorName varchar(100),
Brand int,
PurchasePrice decimal(10,2),
Volume ,
ActualPrice decimal(10,2),
TotalPurchaseQuantity int,
TotalPurchaseDollars decimal(15,2),
Total_sales_Dollers decimal(15,2),
Total_sales_price decimal(15,2),
total_sales_Quantity int,
Total_Excise_Tax decimal(15,2),
Freight_cost decimal(15,2),
GrossProfit decimal(15,2),
ProfitMargin decimal(15,2),
StockTurnover decimal(15,2),
SalestoPurchaseRatio decimal(15,2),
primary key (VendorNumber,Brand)
);""")
```
